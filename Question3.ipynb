# Question 3: Coupled dynamical and kinematic update

```python
import numpy as np
from solver import EulerAngleSolver, EulerRigidBodyDynamics
from plotter import SimulationPlotter
```

Import Data:

```python
# Principal inertias (kg m²)
I1 = 210.0
I2 = 200.0
I3 = 118.0
INERTIA_TENSOR = np.diag([I1, I2, I3])

# Initial conditions
INITIAL_EULER_ANGLES = np.array([0.0, 0.0, 0.0])  # Roll (φ), Pitch (θ), Yaw (ψ)
INITIAL_ANGULAR_VELOCITY = np.array([0.2, 0.15, 0.18])  # ω1, ω2, ω3
INITIAL_CONDITIONS = np.concatenate([INITIAL_EULER_ANGLES, INITIAL_ANGULAR_VELOCITY])

dynamics_handler = EulerRigidBodyDynamics(INERTIA_TENSOR)
```

```python
# Simulation Parameters
T_INITIAL = 0.0
T_FINAL = 30.0
DT = 0.01

solver = EulerAngleSolver(INITIAL_CONDITIONS, T_INITIAL, T_FINAL, DT, dynamics_handler)
time_points, state_history = solver.solve()

print("Simulation complete.")
print(f"Final Euler angles: {state_history[-1, :3]} rad")
print(f"Final angular velocity: {state_history[-1, 3:]} rad/s")
```

```python
# Plot the results
SimulationPlotter.plot_full_state(time_points, state_history, DT)
```
